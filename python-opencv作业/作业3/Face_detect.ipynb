# 人脸检测

```python
# 倒入需要的库
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('001.png') #加载图片
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #转化为RGB通道
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #转换成灰度图
plt.imshow(img_gray, cmap ='gray') #显示一下灰度图
```

```python
detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') #加载opencv官方的人脸数据,生成一个人脸检测器
faces = detector.detectMultiScale(img_gray, scaleFactor = 1.3, minNeighbors = 5,minSize=(30, 30), ) #在灰度图上检测人脸
# scaleFactor。有的人脸离镜头近，会比其他人脸更大。ScaleFactor 对此进行补偿。
# 实测这张图片参数设置在1.2-1.4人脸检测效果最好
# minNeighbors 会对当前其周围有多少物体进行定义
# minsize 每个检测窗口的大小
# 返回的faces包含检测框的左上角顶点的位置和检测框的高和宽，形成一个列表
```

```python
img_detector = img.copy() 
#用copy()将原图复制一份 不能直接赋值给img_detector,这样会将img的地址给img_detecror,起不到将img复制的作用
# 目的是不在原图上画检测框
```

```python
for (x, y, w, h) in faces: #遍历所有的检测框
    cv2.rectangle(img_detector, (x, y), (x + w, y + h), (255, 0, 0), 3) #检测框颜色选择红色，线宽3个像素
plt.imshow(img_detector) #显示结果
```
